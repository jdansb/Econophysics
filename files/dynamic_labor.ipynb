```python
import numpy as np

N = 1000000

count = 0
Ds = []
P=[]
M=[]
L=[]
logs=[]
for _ in range(N):

    # preço específico (sempre positivo)
    psi1 = np.random.normal(loc=100, scale=10)
    psi2 = np.random.normal(loc=100, scale=10)

    # trabalho contido
    L1 = np.random.normal(loc=100, scale=10)
    L2 = np.random.normal(loc=100, scale=10)

    # preços monetários
    M1 = psi1 * L1
    M2 = psi2 * L2

    M.append(M1)
    M.append(M2)
    P.append(psi1)
    P.append(psi2)
    L.append(L1)
    L.append(L2)

    # definir D usando a cesta mais cara e a mais barata
    if M1 > M2:
        D = L1 / L2
    else:
        D = L2 / L1

    Ds.append(D)
    logs.append(np.log(D))

    if D > 1:
        count += 1

Ds = np.array(Ds)
logs = np.array(logs)
print("P(D>1) =", count / N)
print("<D> =", np.mean(Ds))
print("<logs(D)> =c = ", np.mean(logs))
```
