```python
#Vamos conferir se os preços específicos tendem a cair.

import numpy as np


N = 1000000 #número de testes
c = 0    #Vezes que o LT caiu
for x in range(N):
  #Preço Específico
  P1 = np.random.normal(loc=100, scale=10)
  P2 = np.random.normal(loc=100, scale=10)
  #$Labr content das mercadoris
  L1 = np.random.normal(loc=100, scale=10)
  L2 = np.random.normal(loc=100, scale=10)
  #Valor Monetário
  M1=P1*L1
  M2=P2*L2
  if (M1<M2 and L1<L2) or (M2<M1 and L2<L1):
    c+=1


print(c/N)
```
