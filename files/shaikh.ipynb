Based on the paper "[Some universal patterns in income distribution: An econophysics approach](https://onlinelibrary.wiley.com/doi/abs/10.1111/meca.12412)", We will work with the "two-class" approach from econophysics (known as EPTC, or "EconoPhysics Two Class"). The original article argues that labor income follows an exponential distribution (for approximately 97%–99% of the population), whereas property income follows a Pareto distribution.

-  Property income refers to earnings derived from asset ownership rather than the direct provision of labor. This category includes, for example, interest, dividends, rents, royalties, and other income stemming from the ownership of financial, real estate, or intellectual assets.
- In contrast, labor income corresponds to earnings received in exchange for labor, such as wages, fees, and commissions.

The idea is to model the per capita income of several population subgroups as being proportional to the product of two variables:

-  GDP per capita;
- The inverse of the Gini coefficient, that is, if $G$ denotes the Gini coefficient, then the proportionality is given by $g=1-G$.


Let us denote the population and total income of the $i$-th fraction by $X_{i}$ and $Y_{i}$, respectively. Then, for the total population, we have

\begin{equation}
X=\sum^{n}_{i}X_{i},\quad Y=\sum^{n}_{i}Y_{i}.
\end{equation}

The per capita income is therefore simply

\begin{equation}
\overline{y}=\frac{Y}{X}.
\end{equation}

The cumulative values are given by

\begin{equation}
X\left(x\right)=\sum^{x}_{i=1}X_{i},\quad
Y\left(x\right)=\sum^{x}_{i}Y_{i},\quad
\overline{y}\left(x\right)=\frac{Y\left(x\right)}{X\left(x\right)}.
\end{equation}

Similarly, the cumulative population and income shares are

\begin{equation}
x=\sum^{x}_{i=1}\frac{X_{i}}{X}=\frac{X\left(x\right)}{X},\quad
y\left(x\right)=\sum^{x}_{i}\frac{Y_{i}}{Y}=\frac{Y\left(x\right)}{Y}.
\end{equation}



Then, the ratio between the per capita income of the bottom population share and the national average income is equal to the ratio between the cumulative income share of that population segment and its corresponding share of the total population. That is,

\begin{equation}
IR\left(x\right)
=\frac{\overline{y}\left(x\right)}{\overline{y}}
=\frac{\frac{Y\left(x\right)}{X\left(x\right)}}{\frac{Y}{X}}
=\frac{Y\left(x\right)}{Y}\frac{X}{X\left(x\right)}
=y\left(x\right)\frac{1}{x}.
\end{equation}

Or, more simply,

\begin{equation}
IR\left(x\right)=\frac{y\left(x\right)}{x}.
\end{equation}

This quantity can also be interpreted as a measure of inequality. However, unlike the Gini coefficient, which captures the Lorenz curve as a whole, $IR(x)$ characterizes only a single point on the curve.

We now consider the Lorenz curve as a combination of two distributions. Let us define:

- $y'\left(x\right)$: the cumulative income share arising from the exponential component of the Lorenz curve;
- $G'$: the Gini coefficient associated with this exponential component;
- $f$: the fraction of total income contained in the Pareto component, that is, the share of total income corresponding to property income;
- A step-function approximation,
\begin{equation}
\theta\left(x-1\right)=
\begin{cases}
0, & x< 1,\quad \text{(exponential component)},\\
1, & x=1.
\end{cases}
\end{equation}

As I will discuss later, the exact meaning of the term "components" (or "sections") in this context is not entirely clear. It is not explicitly stated whether they represent a partition of the population into distinct groups or a decomposition of income according to its source. Under the latter interpretation, the same segment of the population - particularly individuals in the upper tail of the income distribution - may simultaneously receive a labor-income component, described by the exponential distribution, and a property-income component, described by the Pareto distribution. In this case, the total income of a given population group would be given by the sum of these two components, rather than by assigning each group exclusively to one of them.


The Lorenz curve is then written as

\begin{equation}
y\left(x\right)=y'\left(x\right)\left(1-f\right)+\theta\left(x-1\right).
\end{equation}

For $x< 1$, this reduces to

\begin{equation}
y\left(x\right)=y'\left(x\right)F,
\end{equation}

where $F=1-f$ denotes the fraction of total income contained in the exponential component. Since $0\leq y'\left(x\right)\leq1$, multiplying by $F$ simply rescales the cumulative income share to the appropriate fraction of total income. By construction,

\begin{equation}
y'\left(1\right)=1,
\end{equation}

that is, when the entire population is considered, the exponential component accounts for all of its own income. However, this represents only a fraction $F$ of the total income.

For $x=1$, the above expression gives

\begin{equation}
y\left(1\right)=F+1=2-f,
\end{equation}

which is inconsistent with the normalization condition of the Lorenz curve, namely $y(1)=1$. Therefore, it seems more appropriate to write

\begin{equation}
y\left(x\right)
=
y'\left(x\right)\left(1-f\right)
+
f\,\theta\left(x-1\right).
\end{equation}

Since $\theta(x-1)=0$ for $x< 1$, this modification leaves the expression unchanged over the exponential region. At $x=1$, however,

\begin{equation}
y\left(1\right)
=
1-f+f
=
1,
\end{equation}

thus satisfying the required normalization. The Gini coefficient is likewise written as

\begin{equation}
1-G=\left(1-G'\right)\left(1-f\right),
\end{equation}

or, equivalently,

\begin{align}
G
&=1-\left(1-G'\right)\left(1-f\right) \\
&=1-\left(1-f-G'+G'f\right) \\
&=G'+f\left(1-G'\right).
\end{align}

To understand this result, we should recall the geometric interpretation of the Gini coefficient. Geometrically,

\begin{equation}
G=\frac{A}{A+B},
\end{equation}

where $A$ is the area between the line of perfect equality and the Lorenz curve, while $B$ is the area under the Lorenz curve.

If both axes are normalized so that $0\leq x,y\leq1$, the area under the line of perfect equality is

\begin{equation}
A+B=\frac{1}{2},
\end{equation}

which implies

\begin{equation}
G=2A.
\end{equation}

Since

\begin{equation}
2A=1-2B,
\end{equation}

it follows that

\begin{equation}
G=1-2B.
\end{equation}

Equivalently,

\begin{equation}
\frac{1-G}{2}=B.
\end{equation}

Because $B$ is precisely the area under the Lorenz curve, the Gini coefficient can be obtained by integrating the Lorenz curve over the interval $[0,1]$:

\begin{equation}
\frac{1-G}{2}
=
\int_{0}^{1}y(x)\,dx,
\end{equation}

or, equivalently,

\begin{equation}
G
=
1-2\int_{0}^{1}y(x)\,dx.
\end{equation}

Using

\begin{equation}
y(x)
=
y'(x)\left(1-f\right)
+
f\,\theta(x-1),
\end{equation}

there is an important conceptual issue that should be addressed before proceeding. We have defined $y'(x)$ as the cumulative income share associated with the exponential component of the Lorenz curve, and $G'$ as the Gini coefficient corresponding to that same component. However, the precise meaning of the term "exponential component" is not entirely clear.

One possible interpretation is that it refers to the portion of the population whose income follows the exponential distribution, corresponding to the interval $x\in[0,1)$. An alternative interpretation is that it represents the exponential contribution to the cumulative income of the entire population, defined over $x\in[0,1]$, with the understanding that at $x=1$ the cumulative income contains both the exponential contribution and the Pareto contribution.

This distinction, however, does not affect the calculation of $G'$, since there is no mathematical difference between integrating over $[0,1)$ and over $[0,1]$: the single point $x=1$ has zero measure and therefore contributes nothing to the integral.

Furthermore, it is worth noting that $y'(x)$ is already a normalized Lorenz curve, whereas the total Lorenz curve is obtained simply by rescaling the vertical coordinate according to

\begin{equation}
y(x)
=
y'(x)\left(1-f\right)
+
f\,\theta(x-1).
\end{equation}

Since the horizontal coordinate $x$ is never renormalized, this construction suggests that the expression "exponential component" does not refer to a subpopulation, but rather to the exponential contribution to the income distribution of the entire population. Under this interpretation, at $x=1$ the cumulative income simultaneously contains the exponential contribution and the Pareto contribution introduced through the Heaviside step function. This interpretation is further supported by the form in which we have chosen to write the Lorenz curve:

\begin{equation}
y\left(x\right)
=
\underbrace{y'\left(x\right)\left(1-f\right)}_{\text{exponential component}}
+
\underbrace{f\theta\left(x-1\right)}_{\text{Pareto component}}.
\end{equation}

For $x< 1$, only the exponential component contributes, whereas at $x=1$ we have

\begin{equation}
y\left(1\right)
=
\underbrace{\left(1-f\right)}_{\text{exponential component}}
+
\underbrace{f}_{\text{Pareto component}}.
\end{equation}

Thus, at the endpoint of the Lorenz curve, the cumulative income receives contributions from both components. With this interpretation in mind, we may proceed with the calculation of the Gini coefficient:

\begin{align}
G
&=1-2\int_{0}^{1}y(x)\,dx\\
&=1-2\int_{0}^{1}\left[y'(x)\left(1-f\right)+f\theta(x-1)\right]dx\\
&=\left(1-2\left(1-f\right)\int_{0}^{1}y'(x)\,dx\right)
-2f\int_{0}^{1}\theta(x-1)\,dx\\
&=\left(1-2\left(1-f\right)\left(\frac{1-G'}{2}\right)\right)
-2f\int_{0}^{1}\theta(x-1)\,dx\\
&=\left(1-\left(1-G'-f+fG'\right)\right)
-2f\int_{0}^{1}\theta(x-1)\,dx\\
&=\left(G'+f\left(1-G'\right)\right)
-2f\int_{0}^{1}\theta(x-1)\,dx,
\end{align}

where we have used

\begin{equation}
\frac{1-G'}{2}
=
\int_{0}^{1}y'(x)\,dx.
\end{equation}

Now perform the change of variables

\begin{equation}
u=x-1,
\qquad
du=dx.
\end{equation}

The integration limits become

\begin{equation}
x=0\;\Rightarrow\;u=-1,
\qquad
x=1\;\Rightarrow\;u=0,
\end{equation}

so that

\begin{equation}
\int_{0}^{1}\theta(x-1)\,dx
=
\int_{-1}^{0}\theta(u)\,du.
\end{equation}

The Heaviside step function is conventionally defined as

\begin{equation}
\theta(x)
=
\begin{cases}
1, & x\geq0,\\
0, & x<0.
\end{cases}
\end{equation}

This is precisely the case here. The corresponding antiderivative of the Heaviside step function is

\begin{equation}
\int_{-\infty}^{x}\theta(u)\,du
=
x\,\theta(x)
=
\max\{0,x\}.
\end{equation}

Since $\theta(u)=0$ throughout the interval $[-\infty,-1)$, we may write

\begin{equation}
\int_{-1}^{0}\theta(u)\,du
=
\int_{-\infty}^{0}\theta(u)\,du
=
\max\{0,0\}
=
0.
\end{equation}

More explicitly,

\begin{align}
\int_{-\infty}^{0}\theta(u)\,du
&=
\int_{-\infty}^{-1}\theta(u)\,du
+
\int_{-1}^{0}\theta(u)\,du,\\
\int_{-\infty}^{0}\theta(u)\,du
-
\int_{-\infty}^{-1}\theta(u)\,du
&=
\int_{-1}^{0}\theta(u)\,du,\\
\max\{0,0\}
-
\max\{0,-1\}
&=
\int_{-1}^{0}\theta(u)\,du,\\
0
&=
\int_{-1}^{0}\theta(u)\,du.
\end{align}

Therefore, the second integral vanishes, leaving

\begin{equation}
G
=
G'
+
f\left(1-G'\right).
\end{equation}

Now suppose that we consider only the exponential component of the Lorenz curve. In this case,

\begin{equation}
y(x)
=
y'(x)\left(1-f\right).
\end{equation}

Dividing both sides by $1-G$, and using

\begin{equation}
1-G
=
\left(1-G'\right)\left(1-f\right),
\end{equation}

we obtain

\begin{align}
y(x)
&=
y'(x)\left(1-f\right),\\
\frac{y(x)}{1-G}
&=
\frac{y'(x)\left(1-f\right)}
{\left(1-G'\right)\left(1-f\right)},\\
\frac{y(x)/x}{1-G}
&=
\frac{y'(x)/x}{1-G'}.
\end{align}

If we define

\begin{equation}
a(x)
\equiv
\frac{y'(x)/x}{1-G'},
\end{equation}

and recall that

\begin{equation}
IR(x)
=
\frac{\overline{y}(x)}{\overline{y}}
=
\frac{y(x)}{x},
\end{equation}

then

\begin{align}
\frac{y(x)/x}{1-G}
&=
\frac{y'(x)/x}{1-G'}\\
\frac{\overline{y}(x)/\overline{y}}{1-G}
&=
a(x).
\end{align}

Therefore, the per capita income of any cumulative population share $x$ within the exponential component is

\begin{equation}
\overline{y}(x)
=
a(x)\left(1-G\right)\overline{y}.
\end{equation}



## Exponential Distribution

Since Shaikh's article discusses several results obtained in the paper "[Temporal evolution of the "thermal" and "superthermal" income classes in the USA during 1983-2001](https://iopscience.iop.org/article/10.1209/epl/i2004-10330-3)", it is worthwhile to briefly review some of its main findings.

In that work, the authors show that the income distribution in the United States exhibits a well-defined two-class structure:

- The vast majority of the population (97-99\%) belongs to the lower-income class and follows an exponential income distribution that remains remarkably stable over time.
- The upper-income class (1-3\% of the population) follows a power-law (Pareto) distribution whose parameters vary significantly over time, closely tracking the rise and fall of the stock market.


Let us introduce the probability density function $P(r)$, such that the probability of observing an income in the interval $(r,r+dr)$ is given by $P(r)\,dr$. The complementary cumulative distribution function (CCDF), that is, the probability of observing an income greater than $r$, is then

\begin{equation}
C(r)
=
\int_{r}^{\infty}P(r')\,dr'.
\end{equation}

where

\begin{equation}
C(0)=1,
\end{equation}

that is, every individual has an income $r>0$.

Two important distributions are commonly considered.

- The exponential (Boltzmann-Gibbs) distribution:


\begin{equation}
P(r)
\propto
\exp\left(-\frac{r}{T}\right),
\qquad
T=\langle r\rangle
=
\int_{0}^{\infty}r'P(r')\,dr',
\end{equation}

where $T$ is a parameter equal to the average income, commonly referred to as the *income temperature*. For an exponential distribution, the complementary cumulative distribution is also exponential,

\begin{equation}
C(r)
\propto
\exp\left(-\frac{r}{T}\right).
\end{equation}

- A power-law (Pareto) distribution:

\begin{equation}
P(r)
\propto
\frac{1}{r^{\alpha+1}},
\qquad
C(r)
\propto
\frac{1}{r^{\alpha}}.
\end{equation}

The Lorenz curve is defined parametrically in terms of the coordinates $x(r)$ and $y(r)$:


\begin{equation}
x(r)
=
\int_{0}^{r}P(r')\,dr',
\end{equation}

which represents the fraction of the population with income below $r$.

\begin{equation}
y(r)
=
\frac{\displaystyle\int_{0}^{r}r'P(r')\,dr'}
{\displaystyle\int_{0}^{\infty}r'P(r')\,dr'},
\end{equation}

which represents the total income earned by that fraction of the population, expressed as a fraction of the total income of the entire system.


For a purely exponential distribution,

\begin{equation}
P(r)
=
k\exp\left(-\frac{r}{T}\right),
\end{equation}

we first determine the normalization constant:

\begin{align}
1
&=
\int_{0}^{\infty}P(r')\,dr'\\
&=
\int_{0}^{\infty}
k\exp\left(-\frac{r'}{T}\right)\,dr'\\
&=
-Tk
\left[
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{\infty}\\
&=
Tk.
\end{align}


where we have made the substitution

\begin{equation}
-\frac{r'}{T}=u,
\end{equation}

so that

\begin{equation}
dr'=-T\,du,
\end{equation}

from which it follows that

\begin{equation}
T=k^{-1}.
\end{equation}

Next, let us compute

\begin{equation}
\langle r\rangle
=
\int_{0}^{\infty}r'P(r')\,dr'
=
k\int_{0}^{\infty}r'
\exp\left(-\frac{r'}{T}\right)dr'.
\end{equation}

Applying integration by parts,

\begin{equation}
\int u\,dv
=
uv-\int v\,du,
\end{equation}

and choosing

\begin{equation}
u=r',
\qquad
dv=\exp\left(-\frac{r'}{T}\right)dr',
\end{equation}

we obtain

\begin{equation}
du=dr',
\qquad
v=-T\exp\left(-\frac{r'}{T}\right).
\end{equation}

Therefore,

\begin{align}
\langle r\rangle
&=
-kT
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{\infty}
-Tk
\int_{0}^{\infty}
\exp\left(-\frac{r'}{T}\right)dr'\\
&=
-kT
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{\infty}
+
T^{2}k.
\end{align}

For the first term,

\begin{align}
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{\infty}
&=
\lim_{r'\rightarrow\infty}
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]
-
\lim_{r'\rightarrow0}
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]\\
&=
\lim_{r'\rightarrow\infty}
\left[
\frac{r'}
{\exp\left(\frac{r'}{T}\right)}
\right]
-
0\cdot1\\
&=
\lim_{r'\rightarrow\infty}
\left[
\frac{1}
{\frac{1}{T}\exp\left(\frac{r'}{T}\right)}
\right]\\
&=
0.
\end{align}

Since the limit has the indeterminate form $\infty/\infty$, we have applied L'Hôpital's rule,

\begin{equation}
\lim_{x\rightarrow a}
\frac{f(x)}{g(x)}
=
\lim_{x\rightarrow a}
\frac{f'(x)}{g'(x)}.
\end{equation}

Hence,

\begin{equation}
\langle r\rangle
=
T^{2}k
=
\frac{T^{2}}{T}
=
T.
\end{equation}

As expected,

\begin{equation}
x(r)
=
\int_{0}^{r}P(r')\,dr'
=
k\int_{0}^{r}
\exp\left(-\frac{r'}{T}\right)dr'
=
-Tk
\left[
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{r}
=
-Tk
\left[
\exp\left(-\frac{r}{T}\right)-1
\right].
\end{equation}

Furthermore,

\begin{align}
\int_{0}^{r}r'P(r')\,dr'
&=
-kT
\left[
r'
\exp\left(-\frac{r'}{T}\right)
\right]_{0}^{r}
-Tk
\int_{0}^{r}
\exp\left(-\frac{r'}{T}\right)dr'\\
&=
-kT
\left[
r
\exp\left(-\frac{r}{T}\right)
\right]
-
T^{2}k
\left[
\exp\left(-\frac{r'}{T}\right)-1
\right].
\end{align}

Therefore,

\begin{equation}
y(r)
=
\frac{\displaystyle\int_{0}^{r}r'P(r')\,dr'}
{\displaystyle\int_{0}^{\infty}r'P(r')\,dr'}
=
\frac{
-kT
\left[
r
\exp\left(-\frac{r}{T}\right)
\right]
-
T^{2}k
\left[
\exp\left(-\frac{r'}{T}\right)-1
\right]
}{T}.
\end{equation}

Hence,

\begin{equation}
y(r)
=
-k
\left[
r
\exp\left(-\frac{r}{T}\right)
\right]
-
Tk
\left[
\exp\left(-\frac{r'}{T}\right)-1
\right].
\end{equation}

Solving the expression for $x(r)$ in terms of $r$, we obtain

\begin{align}
x
&=
-Tk
\left[
\exp\left(-\frac{r}{T}\right)-1
\right],\\
\frac{x}{-Tk}+1
&=
\exp\left(-\frac{r}{T}\right),\\
\ln\left(\frac{x}{-Tk}+1\right)
&=
-\frac{r}{T},\\
-T
\ln\left(\frac{x}{-Tk}+1\right)
&=
r.
\end{align}

Substituting this expression into $y(r)$, we obtain

\begin{align}
y(r)
&=
-k
\left[
r
\exp\left(-\frac{r}{T}\right)
\right]
-
kT
\left[
\exp\left(-\frac{r}{T}\right)-1
\right]\\
&=
-k
\left[
-T
\ln\left(\frac{x}{-Tk}+1\right)
\exp\left(
\ln\left(\frac{x}{-Tk}+1\right)
\right)
\right]
-
Tk
\left[
\exp\left(
\ln\left(\frac{x}{-Tk}+1\right)
\right)
-1
\right]\\
&=
-k
\left[
-T
\ln(1-x)(1-x)
\right]
-
kT
\left[
(1-x)-1
\right].
\end{align}

Therefore,

\begin{equation}
y(x)
=
(1-x)\ln(1-x)+x.
\end{equation}

This result was also obtained in the article "[Evidence for the exponential distribution of income in the USA](https://arxiv.org/abs/cond-mat/0008305)". We may verify that

\begin{equation}
y(x=0)
=
\ln(1)(1)+0
=
0,
\end{equation}

and

\begin{equation}
\lim_{x\rightarrow1^-}y(x)
=
1,
\end{equation}

since

\begin{align}
\lim_{x\rightarrow1^-}
\ln(1-x)(1-x)
&=
\lim_{x\rightarrow1^-}
\frac{\ln(1-x)}
{\dfrac{1}{1-x}}\\
&=
\lim_{x\rightarrow1^-}
\frac{\dfrac{1}{1-x}}
{-\dfrac{1}{(1-x)^2}}\\
&=
-\lim_{x\rightarrow1^-}
\frac{(1-x)^2}{1-x}\\
&=
-\lim_{x\rightarrow1^-}(1-x)\\
&=
0.
\end{align}

Here we evaluate the left-hand limit because $x\in[0,1]$, and we apply L'Hôpital's rule since the expression has the indeterminate form $\infty/\infty$.

According to "[Statistical Mechanics of Money, Income, and Wealth: A Short Survey](https://arxiv.org/abs/cond-mat/0211175)", the empirical income distribution (using U.S. data from 1997) is well described by a combination of an exponential distribution and a power-law distribution. Consequently, the authors propose the following modification of the Lorenz curve:

\begin{align}
y(x)
&=
\underbrace{(1-f)y'(x)}_{\text{exponential component}}
+
\underbrace{f\Theta(x-1)}_{\text{Pareto component}}\\
&=
(1-f)
\left[
(1-x)\ln(1-x)+x
\right]
+
f\Theta(x-1).
\end{align}

where

\begin{equation}
\Theta(x-1)
=
\begin{cases}
0, & x< 1,\\
1, & x\geq1,
\end{cases}
\end{equation}

and the factor $(1-f)$ is required to ensure the correct normalization:

\begin{align}
y(x=0)
&=
(1-f)
\left[
\ln(1)(1)+0
\right]
+0
=
0,\\
y(x=1)
&=
(1-f)
\left[
\ln(0)(0)+1
\right]
+f
\approx
1-f+f
=
1.
\end{align}

The factor $(1-f)$ represents the fraction of the total income associated with the exponential component of the distribution. Indeed, for every $x< 1$, we have $\Theta(x-1)=0$, so that the Lorenz curve reduces to the exponential expression multiplied by $(1-f)$. In the limit $x\rightarrow1^{-}$, this component accumulates exactly the fraction $(1-f)$ of the total income. The remaining fraction, $f$, corresponds to the income concentrated in the Pareto tail and is incorporated through the term $f\Theta(x-1)$, producing the discontinuity of the Lorenz curve at $x=1$.

## Individual Gini Coefficient

For the exponential component - recalling that by *exponential component* we mean the exponential contribution to $y(x)$, namely $y'(x)$ - the Gini coefficient is

\begin{align}
G
&=
1-2\int_{0}^{1}y(x)\,dx\\
&=
1-2\int_{0}^{1}
\left[
\ln(1-x)(1-x)+x
\right]dx\\
&=
1
-
2\int_{0}^{1}
\ln(1-x)(1-x)\,dx
-
2\int_{0}^{1}x\,dx\\
&=
1
-
2\int_{1}^{0}
u\ln u\,du

## Household Gini Coefficient

The article also considers the household income distribution. To this end, it derives $P_{2}(r)$ under the assumption that both spouses have incomes distributed according to

The probability distribution of the sum of two or more independent random variables is given by the convolution of their individual distributions. The term convolution arises because the probability density function of the sum of independent random variables is precisely the convolution of their respective probability density functions.

For two independent discrete random variables whose values are integers, the general formula for the distribution of the sum

\begin{equation}
Z=X+Y
\end{equation}

is

\begin{equation}
P(Z=z)
=
\sum_{k=-\infty}^{\infty}
P(X=k)
P(Y=z-k).
\end{equation}

The idea is to compute the probability that the sum of the random variables $X$ and $Y$ is equal to a given value $z$. Since $X$ and $Y$ are independent, for each possible value $X=k$, the probability that the events $X=k$ and $Y=z-k$ occur simultaneously is given by the product

\begin{equation}
P(X=k)P(Y=z-k).
\end{equation}

The quantity $z-k$ is precisely the value that $Y$ must take in order to satisfy the condition

\begin{equation}
X+Y=z.
\end{equation}

Since there are many possible values of $k$, we sum these probabilities over all integers.

If, instead, $X$ and $Y$ are independent continuous random variables and $P_{1}(x)$ denotes their probability density function, then the corresponding expression becomes

\begin{equation}
P_{2}(r)
=
\int_{-\infty}^{+\infty}
P_{1}(r')
P_{1}(r-r')
\,dr'.
\end{equation}

Assuming that

\begin{equation}
P_{1}(x)=0,
\qquad
x< 0,
\end{equation}

that is, the probability density function has support only on non-negative values, meaning that the random variable can assume only non-negative values within the model, we can simplify the limits of the convolution. For the integrand

\begin{equation}
P_{1}(r')P_{1}(r-r')
\end{equation}

to be nonzero, it is necessary that

\begin{equation}
r'\geq0,
\qquad
r-r'\geq0,
\end{equation}

which implies

\begin{equation}
0\leq r'\leq r.
\end{equation}

Therefore, the convolution reduces to

\begin{equation}
P_{2}(r)
=
\int_{0}^{r}

## Implications

Two universal rules are proposed throughout the article. The first states that the per capita income of the poorest 70\% of the population is equal to the GDP per capita adjusted by inequality. This corresponds, in fact, to Sen's welfare measure:

\begin{equation}
\overline{y}(0.7)
=
a(0.7)(1-G)\overline{y}
=
(1-G)\overline{y}.
\end{equation}

Therefore, $(1-G)$ represents the available per capita income relative to the first 70\% of the population, that is, $G$ represents the percentage difference between the GDP per capita (national per capita income) and the per capita income of the poorest 70\%.

Now considering the cumulative income within the exponential component:

\begin{equation}
a(0.7)
=
\frac{y'(x)/x}{1-G'}
=
1,
\end{equation}

which implies

\begin{equation}
y'
=
(1-G')0.7.
\end{equation}

We have that $G'$ gives the difference, in percentage terms, from perfect equality. That is, if $G'=0$, corresponding to perfect equality, then

\begin{equation}
y'=0.7,
\end{equation}

meaning that the poorest 70\% of the population accumulates 70\% of the total income.

The second universal rule states that the average income of the bottom 80\% is equal to $1.1$ times the GDP per capita adjusted by inequality:

\begin{equation}
\overline{y}(0.8)
=
a(0.8)(1-G)\overline{y}
=
1.1(1-G)\overline{y}.
\end{equation}

Let us now analyze the variation of per capita income when $G$ and/or $\overline{y}$ change. Considering

\begin{equation}
a(x)=a_x,
\end{equation}

we have

\begin{align}
\overline{y}_{x}
&=
a_x(1-G)\overline{y}\\
\ln(\overline{y}_{x})
&=
\ln\left(a_x(1-G)\overline{y}\right)\\
\ln(\overline{y}_{x})
&=
\ln(a_x)
+
\ln(1-G)
+
\ln(\overline{y})\\
\frac{d}{dt}\ln(\overline{y}_{x})
&=
\frac{d}{dt}
\left[
\ln(a_x)
+
\ln(1-G)
+
\ln(\overline{y})
\right]\\
\frac{\dot{\overline{y}_{x}}}{\overline{y}_{x}}
&=
\frac{\dot{a_x}}{a_x}
-
\frac{\dot{G}}{1-G}
+
\frac{\dot{\overline{y}}}{\overline{y}}\\
\frac{\dot{\overline{y}_{x}}}{\overline{y}_{x}}
&=
\frac{\dot{a_x}}{a_x}
-
\left(
\frac{G}{1-G}
\right)
\frac{\dot{G}}{G}
+
\frac{\dot{\overline{y}}}{\overline{y}}.
\end{align}

Here we have used the chain rule. If we define

\begin{equation}
\frac{1}{x}\frac{dx}{dt}
=
\frac{\dot{x}}{x}
=
\tilde{x},
\end{equation}

and considering, as done in the article, that $a_x$ is constant, so that

\begin{equation}
\tilde{a}_{x}=0,
\end{equation}

we obtain

\begin{equation}
\tilde{\overline{y}}_{x}
=
\tilde{a}_{x}
-
\left(
\frac{G}{1-G}
\right)
\tilde{G}
+
\tilde{\overline{y}}.
\end{equation}

Now, under what conditions does the factor

\begin{equation}
\left(
\frac{G}{1-G}
\right)
\end{equation}

amplify the effect of the percentage variation of the Gini coefficient on the variation of $\tilde{\overline{y}}_{x}$?

The amplification occurs when

\begin{equation}
\frac{G}{1-G}>1.
\end{equation}

Conversely, when

\begin{equation}
\frac{G}{1-G}< 1,
\end{equation}

the opposite effect occurs, attenuating the contribution of the Gini variation, since this factor multiplies the relative variation of $G$.

We can determine precisely the value of the Gini coefficient for which this transition occurs:

\begin{align}
\left(
\frac{G}{1-G}
\right)
&>1\\
G&>1-G\\
2G&>1\\
G&>\frac{1}{2}.
\end{align}

We observe that, although both a decrease in the Gini coefficient and an increase in national per capita income raise the per capita income of the majority of the population, a percentage increase in GDP (national) per capita will always increase the per capita income of the majority by the same proportion.

On the other hand, a given percentage reduction in inequality, measured by the variation of the Gini coefficient, will have a larger positive impact in countries with

\begin{equation}
G>0.50.
\end{equation}

In other words, the partial elasticity of the per capita income of the majority with respect to a $1\%$ variation in GDP per capita will always be equal to one, whereas the absolute value of the elasticity with respect to a $1\%$ variation in the Gini coefficient will be greater than one when

\begin{equation}
G>0.50,
\end{equation}

and smaller than one when

\begin{equation}
G< 0.50.
\end{equation}

Furthermore, returning to

\begin{equation}
G=G'+(1-G')f,
\end{equation}

and taking $G'$ as a fixed quantity, this directly implies that the Gini coefficient is a positive linear function of the fraction of income derived from property income, $f$.



## Conclusion

In this article, the econophysics "two-class" approach (EPTC) to income distribution was presented. This approach postulates and tests the hypothesis that the overall income distribution is composed of two probability distribution functions: the exponential (Boltzmann-Gibbs) distribution, which characterizes the distribution of labor income, and the Pareto distribution (power law), which characterizes the distribution of property income.

The approach derives a simple exponential approximation for the corresponding global Lorenz curve, which we then use to demonstrate that the per capita income of any fraction $x$ of the lower segment of the population is proportional to the national per capita income "discounted by inequality" through a proportionality coefficient $a(x)$, which is an exclusive function of $x$:

\begin{equation}
\overline{y}(x)
=
a(x)(1-G)\overline{y}.
\end{equation}

The article tests this theoretical result using a broad sample of countries from the WIID database and demonstrates that it is extremely robust, both across countries and over time. These findings lead to two universal rules:

1. The "1.1 Rule" establishes that the per capita income of the poorest 80\% of the population of a country can be calculated by multiplying the inequality-adjusted GDP per capita by $1.1$:

\begin{equation}
\overline{y}(0.8)
=
1.1(1-G)\overline{y}.
\end{equation}

2. For the poorest 70\% of the population, the corresponding coefficient is $1$:

\begin{equation}
\overline{y}(0.7)
=
(1-G)\overline{y}.
\end{equation}


This result implies that the inequality-adjusted GDP per capita is equivalent to the per capita income of this fraction of the population. This latter result provides an alternative to the Sen (1976) approach as a measure of social welfare.

Three main implications can be extracted from our conclusions.

1. For international comparisons, a measure such as the VMI (Vast Majority Income), defined as the per capita income of the bottom 80\% of the population, $ \overline{y}(0.8),$ is preferable to GDP, since VMI combines income levels and inequality into a simple and intuitively monetary statistic: the per capita income of the poorest 80\% of the population.

2. We demonstrate that, although both the growth of average per capita income and the reduction of inequality increase VMI, the reduction of inequality has a proportionally smaller effect on VMI in countries with Gini coefficients below $0.50$.

3. We show that the EPTC approach implies that the Gini coefficient is a simple function of the share of property income in total national income. Considering that the Gini coefficient of the exponential section $G'$ is constant, as observed empirically, the Gini coefficient becomes sensitive to changes in the distribution between labor income and property income, but insensitive to changes in the internal distribution within either of these classes. Since the income of the poorest 97\% of the population is composed predominantly of labor income, this result is consistent with the well-known insensitivity of the Gini coefficient to distributive changes occurring in the middle range of the income distribution.
