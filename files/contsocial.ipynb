```python
import numpy as np
```

```python
A = np.array([[45/5, 0.08,0.01],
              [0.18, 0.21, 0.11],
              [0.00, 0.05, 0.00]])
I = np.eye(3)
M=I-A
L = np.linalg.inv(M)
print(L)
```

```python
v = np.array([[200],[2000],[1808]])
x=np.dot(L, v)
print(x)
```
